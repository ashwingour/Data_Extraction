```python
from requests import get
url = 'https://store.nike.com/us/en_us/pw/jordan-clothing/1mdZofq?ipp=198'
response = get(url)
print(response.text[:])
```

```python
from bs4 import BeautifulSoup

html_soup = BeautifulSoup(response.text, 'html.parser')
type(html_soup)
```

```python
item_containers = html_soup.find_all('div', class_ = 'product-name')
print(type(item_containers))

print(len(item_containers))
```

```python
for i in item_containers:
    print(i)
```

```python
for n in range(0,len(item_containers)):
    list_item = item_containers[n].text
    print(list_item)
```
